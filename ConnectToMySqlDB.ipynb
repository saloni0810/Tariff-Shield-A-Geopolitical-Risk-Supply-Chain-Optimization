```python
import pymysql
import pandas as pd
from sqlalchemy import create_engine
```

```python
# MySQL Configuration

HOST = <HOST>
USER = <USER_NAME>
PASSWORD = <PASSWORD>  
PORT = <PORT>
DB_NAME = <DB_NAME>

# ========================
# 2. Create Database if Not Exists
# ========================
conn = pymysql.connect(
    host=HOST,
    user=USER,
    password=PASSWORD,
    port=PORT
)
cursor = conn.cursor()
cursor.execute(f"CREATE DATABASE IF NOT EXISTS {DB_NAME};")
conn.commit()
conn.close()
print(f"Database '{DB_NAME}' created (if not existed).")
```

```python
# Connect to Database with SQLAlchemy
ENCRYPT_PASSWORD = <PASSWORD>
engine = create_engine(f"mysql+pymysql://{USER}:{ENCRYPT_PASSWORD}@{HOST}:{PORT}/{DB_NAME}")

# Load CSV Files
trade_file = "final_trade_data.csv"
tariff_file = "final_tariff_data.csv"
economic_file = "final_economic_data.csv"

# Read CSVs into pandas DataFrames
trade_df = pd.read_csv(trade_file)
tariff_df = pd.read_csv(tariff_file)
economic_df = pd.read_csv(economic_file)

# Push DataFrames to MySQL Tables
trade_df.to_sql("trade_data", engine, if_exists="replace", index=False)
tariff_df.to_sql("tariff_data", engine, if_exists="replace", index=False)
economic_df.to_sql("economic_data", engine, if_exists="replace", index=False)

print("CSV files imported into MySQL tables successfully!")
```

```python
# ========================
# 6. Test Query
# ========================
from sqlalchemy import text
with engine.connect() as conn:
    result = conn.execute(text("SELECT COUNT(*) FROM trade_data;"))
    print("Rows in trade_data:", list(result)[0][0])
```
